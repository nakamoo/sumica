```python
import requests
import os
import json
import os

%load_ext autoreload
%autoreload 2
import controllers.utils as utils

import matplotlib.pyplot as plt
%matplotlib inline

os.chdir("/home/sean/HAI/main_server/sumica")

def show(impath):
    r = requests.post("http://localhost:5002/extract_features",
                      data={'path': os.path.abspath(impath), 'detection_threshold': 0.5, 'nms_threshold': 0.5})
    print(r.status_code)
    features = json.loads(r.text)
    img = plt.imread(impath)

    fig, ax = plt.subplots(figsize=(36, 24))
    
    ax.imshow(utils.visualize(img, features))
```

```python
show("/home/sean/HAI/main_server/sumica/notebooks/sean/folder2/test1.jpg")
```
